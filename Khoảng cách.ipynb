```python
# Đọc tọa độ
toado = pd.read_excel('/content/Tọa độ CORELAP (Ver2).xlsx')
dic = {}
for i in range(len(list(toado["Máy"]))):
  dic[toado.iloc[i, 0]] = [toado.iloc[i, 1],toado.iloc[i, 2]]
print(dic)
axis = dic
```

```python
import math

#axis = {
 #       'MAL': [1, 3],
  #      'SH': [4, 6],
   #     'LA': [8, 6],
    #    'MC': [3, 5],
     #   'DC': [5, 2],
      #  'EWC': [10, 20],
      #  'GRS': [15, 7],
     #   'MI': [12, 12],
    #}
pair = []
for i in axis:
    for j in axis:
        if i != j:
            pair.append([i, j])
for i in pair:
    for j in pair:
        if (i[0] == j[1] and i[1] == j[0]):
            pair.remove(j)
frequency = {}
for i in pair:
    frequency[str(i[0]+'&'+i[1])] = mch1.loc[i[0], i[1]]
    #print(str(i[0]+'&'+i[1], frequency[str(i[0]+'&'+i[1])])


def distance(a, b):
    x1 = axis[a][0]
    y1 = axis[a][1]
    x2 = axis[b][0]
    y2 = axis[b][1]
    d = math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    return d


s = 0
for i in pair:
    d = distance(i[0], i[1])*(frequency[str(i[0]+'&'+i[1])])
    s += d
print('Tổng khoảng cách di chuyển là:',s,'(cm)')
```
