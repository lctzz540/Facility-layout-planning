```python
#Hàm tính tần suất
def tansuat(A,B):
  dem=0
  for i in range(df.shape[0]):
      if df.loc[i,'Task1']==A and df.loc[i,'Task2']==B or df.loc[i,'Task2']==A and df.loc[i,'Task3']==B or df.loc[i,'Task3']==A and df.loc[i,'Task4']==B or df.loc[i,'Task4']==A and df.loc[i,'Task5']==B or df.loc[i,'Task5']==A and df.loc[i,'Task6']==B or df.loc[i,'Task6']==A and df.loc[i,'Task7']== B or df.loc[i,'Task7']==A and df.loc[i,'Task8']==B or df.loc[i,'Task8']==A and df.loc[i,'Task9']==B or df.loc[i,'Task8']==A and df.loc[i,'Task9']==B or df.loc[i,'Task9']==A and df.loc[i,'Task10']==B:
        dem = dem + df.loc[i,'Quantity']
  return dem
```

```python
#Tính full tần suất
df1 = pd.read_excel('/content/Data quy trinh gia cong_3.xlsx',sheet_name = 'Machines')
df2 = df1['Máy']
pair = []                                                         #Tạo list trống tên "pair"
for i in range(df2.shape[0]):
  for j in range(df2.shape[0]):
    if df2.loc[i, ] != df2.loc[j, ]:                              #Nếu máy tên máy ở hàng i khác tên máy ở hàng j
      pair.append([df2.loc[i, ], df2.loc[j, ]])                   #Thêm cặp máy bên trên vào list pair

serial = 0
for i in range(len(pair)):
  if (tansuat(pair[i][0], pair[i][1]) != 0):      #pair[i][0]: gọi ra tên cặp thứ i và tên đầu tiên; pair[i][1]: gọi ra cũng cặp thứ i và tên thứ 2
    serial += 1
    print(serial, pair[i][0], end="&")
    print(pair[i][1], end=":")
    print(tansuat(pair[i][0], pair[i][1]))
    
```

```python
#Tính và sắp xếp
df1 = pd.read_excel('/content/Data quy trinh gia cong_3.xlsx',sheet_name = 'Machines')
df2 = df1['Máy']
pair = []
frequency = {}
for i in range(df2.shape[0]):
  for j in range(df2.shape[0]):
    if df2.loc[i, ] != df2.loc[j, ]:                              
      pair.append([df2.loc[i, ], df2.loc[j, ]])

serial = 0
for i in range(len(pair)):
  if (tansuat(pair[i][0], pair[i][1]) != 0):
    serial += 1
    frequency[pair[i][0] + '&' + pair[i][1]] = tansuat(pair[i][0], pair[i][1])
    frequency= dict(sorted(frequency.items(), key=lambda item: item[1])) 
print(frequency)
```

```python
#Phân loại mối quan hệ
for i in frequency:
  if frequency[i] >= 900:
    print('A', end =" ")
  elif frequency[i] >= 400 and frequency[i] < 900:
    print('E', end =" ")
  elif frequency[i] >= 100 and frequency[i] < 400:
    print('I', end =" ")
  elif frequency[i] >= 50 and frequency[i] < 100:
    print('O', end =" ")
  else: print('U', end =" ")
  print(i,frequency[i])
```
